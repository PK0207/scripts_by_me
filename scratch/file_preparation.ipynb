```python
import numpy as np
import pandas as pd
import os
from scipy.io import readsav, savemat
import pickle
import ast
```

```python
target_wavelengths = readsav("nicoleModel/V4046Sgr_H2wavelengths.sav")
target_velocity = pd.read_pickle('nicoleModel/V4046Sgr_H2velocity.pkl')
target_fluxes = pd.read_pickle('nicoleModel/V4046Sgr_H2flux.pkl')
target_fluxerr = pd.read_pickle('nicoleModel/V4046Sgr_H2fluxerr.pkl')
```

```python
target_wavelengths
```

```python
lines = [1460.17, 1467.08]
target_wavelengths_RULupi = {key: [] for key in target_wavelengths.keys()}

for line in lines:
    mask = target_wavelengths['labwave']==line
    for i, key in enumerate(target_wavelengths.keys()):
        if i==0:
            target_wavelengths_RULupi[key].append(float(str(target_wavelengths[key][mask])[1:-1]))
        else:
            target_wavelengths_RULupi[key].append(int(target_wavelengths[key][mask]))

with open('target_wavelengths_RULupi.pkl', 'wb') as f:
    pickle.dump(target_wavelengths_RULupi, f)
```

```python
targets = [target_line, target_2]
```

```python
for target in targets:
    for key in ['flux', 'fluxerr', 'velocity']:
        if key=='velocity':
            velocity_str = target[key][:-7]

            # Remove brackets and strip whitespace/newlines
            clean_str = velocity_str.replace('[', '').replace(']', '').strip()

            # Split by any whitespace (space, newline, tabs)
            str_values = clean_str.split()

            # Convert each to float
            target[key] = [float(val) for val in str_values]
        else:
            fstr = target[key][:-23]
            # Remove brackets and strip whitespace/newlines
            clean_str = fstr.replace('[', '').replace(']', '').strip()

            # Split by any whitespace (space, newline, tabs)
            str_values = clean_str.split()

            # Convert each to float
            target[key] = [float(val) for val in str_values]
```
